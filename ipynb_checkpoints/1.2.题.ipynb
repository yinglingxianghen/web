```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

import matplotlib.pyplot as plt
%matplotlib inline
```

1、numpy图片灰度化处理，三种方式

```python
im_data = plt.imread('芝麻.jpg')
plt.imshow(im_data)
```

```python
#第一种方式最大值
im_data1 = im_data.max(axis = 2)
plt.imshow(im_data1,cmap='gray')
```

```python
#第二种方法，求平均值
im_data2 = im_data.mean(axis = 2)
plt.imshow(im_data2,cmap='gray')
```

```python
#第三种方法，加权值
a = [0.299,0.587,0.114]
im_data3 = np.dot(im_data,a)
plt.imshow(im_data3,cmap='gray')
```

2、numpy实现图片的各种反转和马赛克

```python
im_data = plt.imread('芝麻.jpg')
plt.imshow(im_data)
```

```python
# 垂直方向进行反转
c = im_data[::-1]
plt.imshow(c)
```

```python
# 水平方向进行反转
c2 = im_data[:,::-1,:]
plt.imshow(c2)
```

```python
# 颜色反转
c3 = im_data[:,:,::-1]
plt.imshow(c3)
```
